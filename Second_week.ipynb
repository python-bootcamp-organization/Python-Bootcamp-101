# Wednesday: Elif Statements

Using the elif conditional statement

```python
x, y = 5, 10
if x > y: 
    print("x is greater")
elif x < y:
    print("x is less")
```

Checking Multiple Elif Conditions

```python
# checking more than one elif conditional statement
x, y = 5, 10
if x > y:
    print("x is grater")
elif (x+10) < y:
    print("x is less")
elif  (x+5)==10:
    print("equal")
```

Conditionals Within Conditionals

```python
# writing multiple conditionals within each other - multiple block levels
x, y, z = 5, 10 , 5
if x > y:
    print("greater")
elif x <= y:
    if x == z:
        print("x is equal z")    #resulting output
    elif x != z:
        print("x is not equal to z")    # won't get hit
```

If Statements vs. Elif Statements

```python
# testing output of two if statements in a row that are both true
x, y, z = 5, 10, 5
if x < y:
    print("x is less")
if x == z:
    print("x is equal")
```

```python
# testing output of an if and elif statement that are both true
x, y, z = 5, 10, 5
if x < y:
    print("x is less")
elif x == z:
    print("x is equal to z")
```

# Thursday: Else Statements

1) Writing Your First Else Statement

```python
# using an else statement
name = "John"
if name == "Jacob":
    print("Hello Jacob")
else:
    print("Hello {}!".format(name))
```

2) Complete Conditional Statement

```python
# writing a full conditional statement with if, elif, else
name = "John"
if name[0] == "A":
    print("Name starts with an A")
elif name[0] == "B":
    print("Name starts with a B")
elif name[0] == "J":
    print("Name starts with a J")
else: # covers all other possibilities
    print("Name starts with a {}".format(name[0]))
```
